# Mesclagem final


>Realizei aqueles teste sobre como iria ficar o .csv e agora vou direto ao ponto e terminar de exportar e zipar

```python
df_contabeis = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
# Garantir que DATA está como data
df_contabeis['DATA'] = pd.to_datetime(df_contabeis['DATA'], errors='coerce')
```

```python
#  Criar Ano e Trimestre
df_contabeis['Ano'] = df_contabeis['DATA'].dt.year
df_contabeis['Trimestre'] = df_contabeis['DATA'].dt.quarter
```

```python
#  Mesclagem  com cadastro (CNPJ e Razão Social)
df_contabeis = df_contabeis.merge(
    df_cadastro[['REGISTRO_OPERADORA', 'CNPJ', 'Razao_Social']],
    left_on='REG_ANS',
    right_on='REGISTRO_OPERADORA',
    how='left'
)
```

```python

# Tratando valores ausentes 
df_contabeis['CNPJ'] = df_contabeis['CNPJ'].fillna('NAO_ENCONTRADO')
df_contabeis['Razao_Social'] = df_contabeis['Razao_Social'].fillna('NAO_ENCONTRADO')
```

```python
# Garantindo ValorDespesas como número
df_contabeis['ValorDespesas'] = pd.to_numeric(
    df_contabeis['VL_SALDO_FINAL'],
    errors='coerce'
).fillna(0)
```

```python
# Selecionando apenas as colunas exigidas
df_final = df_contabeis[
    ['CNPJ', 'Razao_Social', 'Trimestre', 'Ano', 'ValorDespesas']
]
```

# Tratamento de Erros Pesados

```python
df_final
```

```python
duplicados = df_final.duplicated(subset=['CNPJ', 'Ano', 'Trimestre'], keep=False)
df_final[duplicados].sort_values(['CNPJ', 'Ano', 'Trimestre']).head(20)
```

```python
# Consolidação final para duplicatas 
df_final = (
    df_final
        .groupby(['CNPJ', 'Razao_Social', 'Ano', 'Trimestre'], as_index=False)
        .agg({'ValorDespesas': 'sum'})
)
```

```python
# garantindo que não haja mais erros 
if df_final.duplicated(['CNPJ', 'Ano', 'Trimestre']).any():
    raise ValueError("Ainda existem duplicatas após consolidação!")
```

```python
df_final # ficou com trimestre de 1,2 e 3 
```

```python
df_final = df_final.sort_values(
    by=['CNPJ', 'Ano', 'Trimestre']
).reset_index(drop=True)
df_final
```

```python

#  Exportar CSV final
df_final.to_csv('data/consolidado_despesas.csv', index=False, encoding='utf-8')

print("✅ sucesso")
```
